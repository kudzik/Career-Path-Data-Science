# Wizualizacja danych - Matplotlib 
## https://matplotlib.org/

`Matplotlib` to jedna z najpopularniejszych bibliotek do tworzenia wykresów w Pythonie. Jest podstawowym narzędziem dla wizualizacji danych w ekosystemie Pythona, powszechnie używanym w data science, analityce danych oraz w badaniach naukowych. Oto krótki opis oraz wady i zalety:

## Opis
Matplotlib umożliwia tworzenie statycznych, animowanych i interaktywnych wykresów. Jego główną zaletą jest ogromna elastyczność i wszechstronność, która pozwala tworzyć wykresy w różnych formatach, takich jak linie, wykresy słupkowe, histogramy, wykresy scatter, konturowe i wiele innych.

## Zalety
- Wszechstronność: Można stworzyć niemal każdy typ wykresu.
- Integracja z innymi bibliotekami: Dobrze współpracuje z innymi bibliotekami jak NumPy, Pandas, oraz SciPy.
- Customizacja: Duże możliwości dostosowywania każdego aspektu wykresów.
- Dokumentacja: Bardzo dobra dokumentacja oraz duża społeczność użytkowników, co ułatwia znalezienie pomocy i przykładowych kodów.

## Wady
- Złożoność: Duża liczba funkcji i opcji może być przytłaczająca dla początkujących użytkowników.
- Wydajność: Przy dużych zbiorach danych i skomplikowanych wykresach może być mniej wydajna niż inne, nowocześniejsze biblioteki.
- Estetyka: Domyślne style Matplotlib są często uważane za mniej nowoczesne i estetyczne w porównaniu do nowszych narzędzi jak Seaborn czy Plotly.


## Alternatywy

- Seaborn: Jest oparta na Matplotlib, ale oferuje bardziej estetyczne wykresy i prostsze interfejsy do tworzenia wykresów statystycznych. Jest szczególnie przydatna do wizualizacji danych z Pandas.

- Plotly: Biblioteka do tworzenia interaktywnych wykresów online. Plotly umożliwia tworzenie wykresów, które użytkownicy mogą przeglądać i analizować, co jest świetne do raportów i prezentacji.

- Bokeh: Biblioteka do tworzenia interaktywnych wykresów w przeglądarce. Bokeh jest idealna do wizualizacji dużych zbiorów danych i oferuje wiele funkcji do dostosowywania wykresów.

- Altair: Jest bardziej deklaratywna niż Matplotlib i oferuje prostszy sposób na tworzenie wykresów. Altair jest łatwa w użyciu i umożliwia szybkie tworzenie estetycznych wykresów.

- Folium: Biblioteka do tworzenia wykresów geograficznych na mapach Google. Jest świetna do wizualizacji danych związanych z lokalizacją.

````python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

matplotlib.__version__
````

````python
fig = plt.figure()
_ = plt.plot([0, 1], [0, 3])
````

````python
_ = plt.plot([0, 1], [0, 3], label="line 1", color="red")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Wykres")
_ = plt.legend()
````

````python
_ = plt.plot([0, 1, 2], [0, 3, 2], label="line 1", color="green")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Wykres")
_ = plt.legend()
````

````python
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)
plt.plot(x, x_sin)
plt.plot(x, x_cos)
````

````python
# Tworzenie wielu wykresów

plt.subplot(121)
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(122)
plt.plot([0, 1, 2], [3, 1, 4], "r--")
````

````python
# Wykresy słupkowe

_ = plt.bar(x=[0, 1, 2, 3], height=[1.2, 3.2, 1.2, 5.2])
````

````python
_ = plt.bar(x=["a", "b", "c", "d"], height=[1.2, 3.2, 1.2, 5.2])
````

````python
x = np.random.randn(200)
y = np.random.randn(200)
rozmiar = np.random.randn(200) * 50

plt.scatter(x, y, rozmiar)
````
