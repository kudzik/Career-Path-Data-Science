```python
import numpy as np

grupa = pokemon.groupby('Type 1')
result = grupa['HP'].sum()

# StwÃ³rz nowy DataFrame z wynikami
new_df = pd.DataFrame({'grupa': result.index, 'suma': result.values})
print(new_df)


# Zgrupowanie: ZgrupowaliÅ›my dane wedÅ‚ug kolumny 'grupa'.
# Obliczenie sumy: ObliczyliÅ›my sumÄ™ wartoÅ›ci w kaÅ¼dej grupie.
# Tworzenie nowego DataFrame: StworzyliÅ›my nowy DataFrame z dwiema kolumnami: 'grupa' (indeksy z grupowania) i 'suma' (wyniki obliczeÅ„).
```

```python
pokemony_po_typie = pokemon.groupby('Type 1')
suma_hp = pokemony_po_typie['HP'].sum()
suma_atak = pokemony_po_typie['Attack'].sum()


pd.DataFrame({'grupa': suma_hp.index, 'suma': suma_hp.values, 'suma_atak': suma_atak.values})
```

```python
plt.figure(figsize=(16, 6))
plt.bar(suma_hp.index, suma_hp.values)
```
