# Łączenie danych z kilku źródeł


```python
def fetch_financial_data(company="AMZN"):
    import pandas_datareader.data as web

    return web.DataReader(name=company, data_source="stooq")


apple = fetch_financial_data("AAPL")
amazon = fetch_financial_data("AMZN")
google = fetch_financial_data("GOOGL")
uber = fetch_financial_data("UBER")
```

```python
apple.columns = ["apple_" + x.lower() for x in apple.columns]
amazon.columns = ["amazon_" + x.lower() for x in amazon.columns]
google.columns = ["google_" + x.lower() for x in google.columns]
uber.columns = ["uber_" + x.lower() for x in uber.columns]
```

```python
# apple.head()
# amazon.head()
# google.head()
uber.head()
```

```python
df = pd.concat([apple, amazon, google, uber], axis=1)
pd.set_option("display.float_format", lambda x: "%.2f" % x)
df
```

```python
df_t = df.T
df_t
```

```python
df_t.corr()  ## Współczynniki korelacji
```

```python
df_close = [col for col in df.columns if col.endswith("close")]
df_close
```

```python
df[df_close].corr()
```
