```python
retiail_raw.query("CustomerID == 17850")
```

```python
# SELECT * FROM retial WHERE CustomerID='17850' and UnitPrice > 5;
retiail_raw.query("CustomerID == 17850 and UnitPrice > 5")

retiail_raw[(retiail_raw.CustomerID == 17850) & (retiail_raw.UnitPrice > 5)]
```

```python
# SELECT * FROM retial WHERE CustomerID='17850' or Country='France';
retiail_raw[(retiail_raw.CustomerID == "17850") | (retiail_raw.Country == "France")]
```

```python
# SELECT * FROM retail WHERE InvoiceNo is not null;
retiail_raw[retiail_raw.InvoiceNo.notnull()]
```

```python
# SELECT CustomerID, count(*) FROM retial GROUP BY CustomerID;
retiail_raw.groupby("CustomerID").size()
```

```python
retiail_raw["Revenue"] = retiail_raw.Quantity * retiail_raw.UnitPrice

# SELECT CustomerID, avg(Revenue), count(*) FROM retial GROUP BY CustomerID;
retiail_raw.groupby("CustomerID").aggregate({"Revenue": np.mean, "CustomerID": np.size})
```
