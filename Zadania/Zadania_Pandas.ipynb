```python
# Obliczanie nowych kolumn

import pandas_datareader.data as web

df = web.DataReader("UBER", "stooq")
df.info()

df.head()
```

```python
df.describe()
```

```python
# Obliczanie średniej z open i close

# Definiowanie noej kolumny avg

df["avarage"] = (df.Open + df.Close) / 2.0
df.head()
```

```python
df.sort_index(inplace=True)
df

df = df.head()
df
```

```python
# Metoda shift przesówa dane

df["shifting"] = df["Close"].shift(1)
df

# wynik dzienny
df["daily_change"] = (df["Close"] / df["shifting"]) - 1
df
```

```python
# Usowanie kolumny

del df["High"]
df
```

```python
# minimalna wartośc z kolumny daily_change

min_daily = df.daily_change.min()
min_daily
```

```python
# Histogram

df["daily_change"].hist(bins=100)
```

```python
# Filtrowanie danych

## Wycinanie danych z dodatni stop zwrotu

# df_positive = df["daily_change"] > 0] # maska logiczna
df_positive = df[df["daily_change"] > 0]  # df
df_positive
```

```python
# Średnia pozytywna stopa zwrotu

df.daily_change.mean()
```

```python
df_2024 = df.index >= "2024-01-01"  # maska logiczna
df_2024 = (df.index >= "2024-01-01") & (
    df.index <= "2024-03-30"
)  # maska logiczna przedział
df_sorted = df[df_2024].sort_index()  # maska logiczna zastosowana na df
df_sorted
```

```python
# Dostęp do danych indexu

df_month = df_sorted.index.month
df_month.value_counts()
```
